# Correlation

Correlation is the statistical relation between two variables  
- It is measured between –1 and +1 


types:
- Positive correlation: both variables increase together  
- Negative correlation: one increases while the other decreases  
- Zero correlation: no consistent linear relationship 

## Heatmaps and Correlation

The purpose of a heatmap in correlation is simple:  

- It shows the correlation values between multiple variables in one visual grid.  
- Colors make it easy to spot strong positive, strong negative, or near‑zero relationships at a glance.  
- Instead of reading a table of numbers, you immediately see patterns and clusters of variables that move together or oppose each other.            

```python
cols = ['temp_c', 'humidity', 'o3', 'co', 'no2', 'so2']
corr = df_new[cols].corr()
plt.figure(figsize=(6, 6))
sns.heatmap(corr, annot=True, cmap='magma', center=0)
plt.title('Correlation Heatmap: Weather & Air Quality')
plt.show()
```

- **Pollutant correlations**: PM2.5 and PM10 are strongly correlated, as are CO and NO₂, reflecting linked emission sources.  

**Overall inference:**  
Global weather patterns and air quality are deeply interconnected. This dataset demonstrates how climate variability shapes pollution levels and atmospheric behavior worldwide. While the analysis provides strong diagnostic insights, predictive forecasting would require multi‑year or real‑time datasets to capture evolving trends.
