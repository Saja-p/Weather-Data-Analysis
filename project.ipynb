# Data Cleaning

### cleaning Numerical columns with characters

```python
df.head(1)
```

```python
df.dtypes
```

```python
df['sunrise']=pd.to_datetime(
    df['sunrise']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce')
```

```python
df.head(10)
```

```python
df.dtypes
```

- Changing last_updated object/string datatype to datetime datatype

```python
df['last_updated']= pd.to_datetime(df['last_updated'],
                                   errors='coerce')
```
